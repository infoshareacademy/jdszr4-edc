```python
def corr_color(val):
    if abs(val) == 1.0:
        color = 'grey'
    elif abs(val) > 0.8:
        color = 'red'
    elif abs(val) > 0.5:
        color = 'orange'
    elif abs(val) > 0.3:
        color = 'yellow'
    else:
        color = 'blue'
    return 'color: %s' % color
```

# Analiza edukacji w dziedzienie HR na różnych poziomach,a  zadowolenie z pracy

```python
hr1 = df[(df.EducationField==0) & (df.Education == 1)]
hr2 = df[(df.EducationField==0) & (df.Education == 2)]
hr3 = df[(df.EducationField==0) & (df.Education == 3)]
hr4 = df[(df.EducationField==0) & (df.Education == 4)]
hr5 = df[(df.EducationField==0) & (df.Education == 5)]

#Means calculations
hr1_enivronment = hr1['EnvironmentSatisfaction'].mean()
hr1_job = hr1['JobSatisfaction'].mean()
hr1_relationship = hr1['RelationshipSatisfaction'].mean()
hr1_overall = (hr1_enivronment + hr1_job +hr1_relationship)/3

hr2_enivronment = hr2['EnvironmentSatisfaction'].mean()
hr2_job = hr2['JobSatisfaction'].mean()
hr2_relationship = hr2['RelationshipSatisfaction'].mean()
hr2_overall = (hr2_enivronment + hr2_job +hr2_relationship)/3

hr3_enivronment = hr3['EnvironmentSatisfaction'].mean()
hr3_job = hr3['JobSatisfaction'].mean()
hr3_relationship = hr3['RelationshipSatisfaction'].mean()
hr3_overall = (hr3_enivronment + hr3_job +hr3_relationship)/3

hr4_enivronment = hr4['EnvironmentSatisfaction'].mean()
hr4_job = hr4['JobSatisfaction'].mean()
hr4_relationship = hr4['RelationshipSatisfaction'].mean()
hr4_overall = (hr4_enivronment + hr4_job +hr4_relationship)/3

hr5_enivronment = hr5['EnvironmentSatisfaction'].mean()
hr5_job = hr5['JobSatisfaction'].mean()
hr5_relationship = hr5['RelationshipSatisfaction'].mean()
hr5_overall = (hr5_enivronment + hr5_job +hr5_relationship)/3

#Plots

w=0.2
x = ['below college', 'college', 'bachelor', 'master', 'doctor']
job = [hr1_job, hr2_job, hr3_job, hr4_job, hr5_job]
environment = [hr1_enivronment, hr2_enivronment, hr3_enivronment,hr4_enivronment, hr5_enivronment]
realtionship = [hr1_relationship, hr2_relationship, hr3_relationship, hr4_relationship, hr5_relationship]
overall = [hr1_overall, hr2_overall, hr3_overall, hr4_overall, hr5_overall]

bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]


plt.bar(bar1, job, w, label= 'Job Satisfaction')
plt.bar(bar2, environment, w, label='Environment Satisfaction')
plt.bar(bar3, realtionship, w, label='Relationship Satisfaction')
plt.bar(bar4, overall, w, label='Overall Satisfaction')

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('Field')
plt.xlabel('Education Level')
plt.title('HR education Satisfaction Measurement')
plt.xticks(bar1+w/2, x)
plt.legend()
plt.show()
```
