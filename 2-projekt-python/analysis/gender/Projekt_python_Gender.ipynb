```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df
```

```python
df['Attrition'] = df['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
df
```

```python
# How many employees left the company
left_df = df[df['Attrition'] == 1]
# How many employees stayed in the company
stayed_df = df[df['Attrition'] == 0]
```

```python
print(len(df))         #quantity all
print(len(left_df))      #quantity - people who left
print(len(stayed_df))    #quantity - people who stayed
```

```python
#Job, Environment and Relationship satisfaction analysis grouped by gender.

df_gender = df[['Gender', 'EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction']]
df_gender.sort_values(by=['Gender'])
```

```python
df_gender['Gender'].value_counts()
```

```python
#job satisfaction by gender
plt.figure(figsize = (5,5))
sns.countplot(x = 'Gender', hue = 'JobSatisfaction', data = df)
pass
```

```python
# environment satisfaction by gender
plt.figure(figsize = (5,5))
sns.countplot(x = 'Gender', hue = 'EnvironmentSatisfaction', data = df)
pass
```

```python
#relationship satisfaction by gender
plt.figure(figsize = (5,5))
sns.countplot(x = 'Gender', hue = 'RelationshipSatisfaction', data = df)
pass
```

```python
#distance from home grouped by gender
plt.figure(figsize = (10,10))
sns.countplot(x = 'DistanceFromHome', hue = 'Gender', data = df)
pass
```

```python
#Age of employees grouped by gender 
sns.boxplot(x = 'Age', y = 'Gender', data = df)
pass
```

```python
#Earnings grouped by gender
sns.boxplot(x = 'MonthlyIncome', y = 'Gender', data = df)
pass
```
