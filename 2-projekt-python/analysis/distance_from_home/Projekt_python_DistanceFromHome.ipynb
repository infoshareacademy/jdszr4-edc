```python
def corr_color(val):
    if abs(val) == 1.0:
        color = 'grey'
    elif abs(val) > 0.8:
        color = 'red'
    elif abs(val) > 0.5:
        color = 'orange'
    elif abs(val) > 0.3:
        color = 'green'
    else:
        color = 'blue'
    return 'color: %s' % color
```

```python
x1_corr = x1.corr()
x1_corr = x1_corr.style.applymap(corr_color)
x1_corr
```
