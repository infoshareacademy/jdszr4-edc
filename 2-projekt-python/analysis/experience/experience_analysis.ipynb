```python
def corr_color(val):
    if abs(val) == 1.0:
        color = 'grey'
    elif abs(val) > 0.8:
        color = 'red'
    elif abs(val) > 0.5:
        color = 'orange'
    elif abs(val) > 0.3:
        color = 'yellow'
    else:
        color = 'blue'
    return 'color: %s' % color
```
