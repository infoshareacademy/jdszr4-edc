## Prediction using decision tree

### Sample from test set

```python
def shroom_check(sample, prediction: int, df):
    """
    Function checks if prediction was consistent with the class value from original dataframe
    """
    
    ix = sample.index[0]
    row_ix = df[df.columns[2:]].iloc[ix:ix+1]
    result = row_ix.equals(sample) and prediction == df.iloc[ix]['class_p']
    if result == False:
        print(ix, row_ix.equals(sample), prediction, df.iloc[ix]['class_p'])
    return result
```

```python
s_test = X_test.sample()
s_test
```

```python
prediction = int(model_tree.predict(s_test))
prediction
```

```python
shroom_check(s_test, prediction, df1)
```

#### Checking for 100 records

```python
true_values = 0
false_values = 0
i=0

while i < 100:
    s_test = X_test.sample()
    pred=model_tree.predict(s_test)
    prediction = int(pred)
    check = shroom_check(s_test, prediction, df1)
#     print(check)

    if check == True:
        true_values += 1
    else:
        false_values += 1
    
    i += 1

print(f'True predictions: {true_values} ({100 * true_values / (true_values + false_values)}%)')
print(f'False predictions: {false_values} ({100 * false_values / (true_values + false_values)}%)')
```

#### Testing behaviour of the model when not user does not give all of the values

```python
def make_columns_nan(df, sample, n_randoms: int):
    
    columns_original = set()
    
    for column in df.columns:
        if column[:-2] == 'class':
            continue
        columns_original.add(column[:-2])
    
    columns_original = list(columns_original)
    
    columns_to_nan = set()
    
    for i in range(n_randoms):
        random_column = columns_original[np.random.randint(0, len(columns_original))]
        columns_to_nan.add(random_column)
    
    print(columns_to_nan)
    
    for column in df.columns:
        if column[:-2] in columns_to_nan:
#             print(column)
            sample[column] = np.NaN
    
    return sample
```

```python
g = make_columns_nan(df1, s_test, 4)
g
```
