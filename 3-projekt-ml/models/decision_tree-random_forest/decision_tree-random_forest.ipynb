```python
# import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# import seaborn as sns
# import warnings

# from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import f1_score, accuracy_score, make_scorer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
# from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier

# warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 150)
# %matplotlib inline
```

```python
df1 = pd.read_csv('../../datasets_preprocessed/one_hot_encoded_var.zip')
```

```python
df1.head() 
```

```python
df2 = pd.read_csv('../../datasets_preprocessed/one_hot_encoded_corr.zip')
```

```python
df2.head()
```

# First dataset

```python
X = df1.drop(['class_e', 'class_p'], axis=1)
y = df1['class_p']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=404)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Decision tree

### Model with standard parameters

```python
model_tree = DecisionTreeClassifier()
```

```python
model_tree.fit(X_train, y_train)
print('Decision tree train acc:', model_tree.score(X_train, y_train))
print('Decision tree test acc:', model_tree.score(X_test, y_test))
```

###### Train and test accur

```python
param_random = {'max_depth': [3, 6, None],
              'max_features': list(range(2, X.shape[1], 20))}
print(param_random)
```

```python
random_search = RandomizedSearchCV(model_tree, param_distributions=param_random, n_iter=15, cv=10, verbose=1)
random_search.fit(X_train, y_train)
```

```python
random_search.cv_results_
```

```python
ranks = random_search.cv_results_['rank_test_score']
ranks.sort()
n = 6

print('5 best sets of parameters:')
for i in range(n):
    print(random_search.cv_results_['params'][ranks[i]])
```

## Random forest

### Model with standard parameters

```python
model_forest = RandomForestClassifier()
```

```python
model_forest.fit(X_train, y_train)
print('Random forest train acc:', model_forest.score(X_train, y_train))
print('Random forest test acc:', model_forest.score(X_test, y_test))
```

###### Train and test accuracy equals 1.0. No need for search for better parameters.
